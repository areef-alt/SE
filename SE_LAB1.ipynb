# **USING FILES**

```python
def read_coefficients(file_path):
    with open(file_path, "r") as file:
        content = file.read()
        coefficients = content.split(",")
        return [float(coef) for coef in coefficients]

coefficients = read_coefficients("/content/value.txt")
times = np.linspace(0, 10, 50)
a, b, c = coefficients
temps = a * (times**2) + b * times + c

sns.set_style("darkgrid")
plt.plot(times, temps, label='Temperature', color="limegreen")
plt.xlabel('Time')
plt.ylabel('Temperature')
plt.title('Quadratic Temperature over Time')
plt.legend()
plt.show()

```

# **MULTIPLE SETS USING FILES italicized text**

```python
def read_coefficients(file_path):
    with open(file_path, "r") as file:
        content = file.readlines()
        coefficients = [line.strip().split(",") for line in content]
        return [[float(coef) for coef in line] for line in coefficients]

coefficients_list = read_coefficients("/content/value2.txt")

for i, coefficients in enumerate(coefficients_list):
    times = np.linspace(0, 10, 50)
    a, b, c = coefficients
    temperature = a * (times**2) + b * times + c

    sns.set_style("darkgrid")
    plt.plot(times, temperature, label=f'Temperature (coefficients: {coefficients})')

plt.xlabel('Time')
plt.ylabel('Temperature')
plt.title('Quadratic Temperature over Time')
plt.legend(["set1[34,56,1]","set2[3.4,5.6,7]","set3[0.1,4.0,3]"])
plt.show()


```

# **MUTIPLE SETS USING INPUT**

```python
def quadratic(time):
  a=0.5
  b=0.3
  c=1
  tempreture=a*(time**2)+b*(time)+c
  return tempreture

def quadratic2(time):
  a=34
  b=67
  c=1
  tempreture=a*(time**2)+b*(time)+c
  return tempreture

def quadratic3(time):
  a=100
  b=200
  c=23
  tempreture=a*(time**2)+b*(time)+c
  return tempreture


times1=np.linspace(0,10,100)
times2=np.linspace(0,20,100)
times3=np.linspace(0,30,100)

temp_val1=quadratic(times1)
temp_val2=quadratic(times2)
temp_val3=quadratic(times3)
sns.set_style("darkgrid")
plt.plot(times1, temp_val1, label='Temperature',color="red")
plt.plot(times1, temp_val2, label='Temperature',color="royalblue")
plt.plot(times1, temp_val3, label='Temperature',color="green")
plt.xlabel('Time')
plt.ylabel('Temperature')
plt.title('Quadratic Temperature over Time')
plt.legend(["set1[0.5,0.3,1]","set2[34,67,1]","set3[100,200,23]"])
plt.show()


```
